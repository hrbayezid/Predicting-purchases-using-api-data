```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv("databangladeshi_users.csv")
df.head()
```

```python
df.describe()
```

```python
sns.countplot(x="items", data=df)
plt.show()
```

```python
sns.histplot(x="purchases", data=df ,kde= True)
plt.show()
```

```python
sns.histplot(x="average", data=df ,kde= True)
plt.show()
```

```python
sns.boxplot(x="average",data=df)
plt.show()
```

```python
sns.countplot(x="items" , data=df)
plt.show()
```

```python
sns.barplot(x="location",y="purchases",data=df,hue="purchases",errorbar=('ci', False))
plt.show()
```

```python
sns.scatterplot(x= "lastseen",y="purchases",hue="purchases",data=df)
plt.show()
```

```python
sns.barplot(x="loyalty",y= "purchases",data=df)
plt.show()
```

```python
sns.heatmap(data = df.corr(), annot=True)
plt.show()
```
